```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import os
```

## 1. English

```python
#move to Data folder
os.chdir("../Data")
```

```python
#get revision and outlinks dict
with open("00_en_as_rev_outlinks_m.pkl", 'rb') as pickleFile:
    outlinks = pickle.load(pickleFile)

#get time and revision dict
with open("00_en_as_rev_time_m.pkl", 'rb') as pickleFile:
    times = pickle.load(pickleFile)
```

```python
#get number of outlinks through time
quantity_dict = {key: len(value) for key, value in outlinks.items()}
```

```python
#get rid of the 0s
quantity_dict_clean = {x:y for x,y in quantity_dict.items() if y!=0}

#only run once
times.pop('2013-07-31')
times_clean = times
times_clean.pop('2018-03-31')
```

```python
x = list(times.index)
y = quantity_dict_clean.values()

plt.ylabel('Size Through Time') 
plt.title("'Arab Spring' Number of Outlinks English")

plt.plot(x, y)

#move the dir back
os.chdir("../01_Salience")
plt.savefig('01_Visuals/numOfOutlinks_AS_en.pdf')

plt.show()
```
